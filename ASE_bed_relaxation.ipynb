```python
def divuhplot(file,title):
    amr = amrio.load(filename)
    x,y,h = amrio.readBox2D(amr,lev,lo,hi,'thck',1)
    x,y,r = amrio.readBox2D(amr,lev,lo,hi,'topg',1)
    x,y,divuh = amrio.readBox2D(amr,lev,lo,hi,'divuh',1)
    x,y,a = amrio.readBox2D(amr,lev,lo,hi,'divuho',1)
    x,y,ux = amrio.readBox2D(amr,lev,lo,hi,'xVelb',1)
    x,y,uy = amrio.readBox2D(amr,lev,lo,hi,'yVelb',1)
    x,y,uxo = amrio.readBox2D(amr,lev,lo,hi,'xVelo',1)
    x,y,uyo = amrio.readBox2D(amr,lev,lo,hi,'yVelo',1)
    x,y,uc = amrio.readBox2D(amr,lev,lo,hi,'velc',1)
    amrio.free(amr)
    
    hab = h + rhoo / rhoi * r
    u = (ux**2 + uy**2)**0.5
    uo = (uxo**2 + uyo**2)**0.5
    umis = uc*(u-uo)

    plt.figure(1,figsize=[12,8])
    
    
    plt.subplot(221)
    plt.pcolormesh(x/1.0e3,y/1.0e3,uc*(a-divuh),norm=divuhlim)
    plt.colorbar()
    plt.contour(x/1.0e3,y/1.0e3,hab, [1] )
    plt.title(r'$M_b - \nabla . (\vec{u}h)$')
    plt.xlabel("x (km)")
    plt.xlabel("y (km)")
    
    plt.subplot(222)
    plt.pcolormesh(x/1.0e3,y/1.0e3,umis,norm=umislim)
    plt.colorbar()
    plt.contour(x/1.0e3,y/1.0e3,hab, [1] )
    plt.title(r'$|\vec{u}|-|\vec{u}|_o$')
    plt.xlabel("x (km)")
    plt.xlabel("y (km)")    
    
    
    plt.subplot(223)
    plt.pcolormesh(x/1.0e3,y/1.0e3,h-h0,norm=dhlim)
    plt.colorbar()
    plt.contour(x/1.0e3,y/1.0e3,hab, [1] )
    plt.title(r'$h - h_0)$')
    plt.xlabel("x (km)")
    plt.xlabel("y (km)")
    
    plt.subplot(224)
    plt.pcolormesh(x/1.0e3,y/1.0e3,r,norm=topglim)
    plt.colorbar()
    plt.contour(x/1.0e3,y/1.0e3,hab, [1] )
    plt.title(r'$r$')
    plt.xlabel("x (km)")
    plt.xlabel("y (km)")  
    
    
    plt.show()
```

```python
def norms(filename):
    #return L2-norms of a-div(uh) and h-ho
    amr = amrio.load(filename)
    x,y,h = amrio.readBox2D(amr,lev,lo,hi,'thck',1)
    x,y,hc = amrio.readBox2D(amr,lev,lo,hi,'thkc',1)
    x,y,uc = amrio.readBox2D(amr,lev,lo,hi,'velc',1)
    x,y,divuh = amrio.readBox2D(amr,lev,lo,hi,'divuh',1)
    x,y,a = amrio.readBox2D(amr,lev,lo,hi,'divuho',1)
    amrio.free(amr)
    dx2 = 4.0e+3 / (2.0**lev)    
    
    return dx2/1.0e9 * np.sqrt(np.sum( uc*(a-divuh)**2 )), dx2/1.0e9 * np.sqrt(np.sum( hc*((h-h0)**2) ))
    
sq =  np.arange(0,128,8)
L = len(sq)
p = np.zeros(L)
q = np.zeros(L)
dirname="data/BISICLES-Antarctica2013/control/ase/mass-balance/test13/"
for i in np.arange(0,L):
    filename = dirname + "ase-test13-ctrl-2lev-outer.%06d.2d.hdf5" % sq[i]
    p[i],q[i] = norms(filename)
    
plt.plot(p,q,'r-')

j = 6
plt.plot(p[j],q[j],'ro')
plt.show()
```
