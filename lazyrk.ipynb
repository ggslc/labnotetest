# Lazy Euler and Lazy RK2

This is really a test of the Ipython notebook, mixing inline text and LaTeX, python code and plots. We are solving an ODE 

$\frac{dp}{dt} + p^2 = 0 \qquad:\qquad  p(0) = 1$

which has an analytic solution:

$p(t) = 1/(t+1)$.

The idea is that there is a hard part of the rhs ($p^{2-s}$) which we only want to evaluate occasionally (like u as a solution to the stress equations), and an easy part $p^s$ which we can evaluate, and must do so for stability (like $\nabla.(\vec{u}h)$ in the thickness transport - we need to evaluate update h every CFL timestep)

Question, is there any value in constructing a kind of 'lazy' Runge-Kutta RK2 method for BISICLES, where we compute $vec{u}$, advance h by N timesteps updating onlh $h$ in $\nabla.(\vec{u}h)$ , recompute $\vec{u}$ given $h$, set $h$ to its initial value, and then advance by $2N$ timesteps given the new $\vec{u}$.

In other words, is the lazy RK2 any more acurate than Euler?

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt 

def lazyplots(split):
    
    def lazyeuler(p,dt,n):
        p0 = p
        f = p**(2-split)
        for i in range(0,(2*n)):
            p = p - dt * f * p**split
        return p,2*n*dt


    def lazyrk(p,dt,n):
   
        p0 = p
        f = p**(2-split)
        for i in range(0,n):
            p = p - dt * f * p**split
        f = p**(2-split)
        p = p0
        for i in range(0,(2*n)):
            p = p -dt * f * p**split
        return p,2*n*dt

    def odesol(ninner,lazyf):
        ntot = 320
        dt = 0.01
        nouter = ntot/ninner
        p = np.zeros(nouter+1)
        p[0] = 1
        t = np.zeros(nouter+1)
        t[0] = 0
        for n in range(1,nouter+1):
            p[n],dtn = lazyf(p[n-1],dt,ninner)
            t[n] = t[n-1] + dtn
    
        return p,t

    plt.figure()

    M = 5
    ee = np.zeros(M)
    er = np.zeros(M)
    N = np.zeros(M,dtype=np.int)

    for i in range(0,M):

        N[i] = 2**(i+1)
        pe,t = odesol(N[i],lazyeuler)
        pr,tr = odesol(N[i],lazyrk)
        plt.plot(t,pe,'r')
        plt.plot(tr,pr,'b')
        a = 1.0/(t+1.0) #analytic solution
        ee[i] = max(abs(pe - a))
        er[i] = max(abs(pr - a))
    
    
    plt.plot(t,a,'g')
    plt.xlabel('t')
    plt.ylabel('p')
    
    
    plt.figure()
    plt.plot(N,ee)
    plt.plot(N,er)

    plt.loglog()
    plt.xlabel('N')
    plt.ylabel('error')


    
```

When split = 0, we simply get the ordinary Euler and RK2 methods (with timestep 2*n*dt), and so we see first order convergence for Euler and second order for RK2

```python
lazyplots(0.0)
```

But for split > 0, lazy RK2 is not equivalent to RK2, and so does not produce second order convergence - but for large N, it is approximately second order

```python
lazyplots(0.5)
```
