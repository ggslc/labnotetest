# MISMIP+ experiments

The MISMIP+ expriments are to be designed by me (S L Cornford), Hilmar (Gudmundsson), with approval from Frank Pattyn and Gael Durand. We agreed at EGU 2014 and IGS 2014 to base the expriments on Hilmar's channel geoemtry in TC 2012 vol 6(p 1497) http://www.the-cryosphere.net/6/1497/2012/. 

The topography $r$ has retrograde and prograde slopes and steep channel walls, and the channel width $w_c$ is a natural parameter to vary in steady-state expriments of the sort described in Hilmar's paper.

$r = B_x(x) + B_y(y)$

where

$B_x(\tilde{x}) = B_0 - 2184.8 \tilde{x}^2 + 1031.72 \tilde{x}^4 - 151.72 \tilde{x}^6$ 

where 

$\tilde{x} = x / 750 {\rm km}$, $0 < x < 1800 {\rm km}$, $0 < y < 240 {\rm km}$

$B_y = d_c / (1 + \exp(-2(y-w_c)/f_c) + d_c / (1 + \exp(2(y-w_c)/f_c)$

Both Ua and Elmer/Ice carried out the steady state experiment and found a stable equilibium on a retrograde slope in the channel (for narrow channels)

For the MISMIP+ experiments, we want to study melt perturbations, so one channel width should be adequate. The parameters $B_0 = 300.0, f_c = 500 {\rm km},  d_c = 1.0{\rm km}, w_c = 40 {\rm km}$ give us a channel 80 km wide



## BISICLES implementation

-Put problem in BISICLES/examples/MISMIPNG

-Used python interface to provide topography etc

```python
#snippet from ghgChannel.py

import math as m

Bo = 300.0
fc = 5.0e+3
dc = 1.0e+3
wc = 40.0e+3

def Bx(x):
    xx = x/750.0e3
    xx2 = xx*xx
    xx4 = xx2*xx2
    xx6 = xx4*xx2
    return Bo - 2184.8 * xx2 + 1031.72 * xx4 - 151.72 * xx6 

def By(y):
    return dc/(1.0+m.exp(-2.0*(y-wc)/fc)) + dc/(1.0+m.exp(2.0*(y+wc)/fc))

def topography(x,y):
    y = y - 120e+3
    topography = Bx(x) + By(y)
    
    return topography
```

```python

#plot the topography
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

dx = 10.0e+3
x = np.arange(0.0+dx/2,1500.0e+3 ,dx)
y = np.arange(0.0 + dx/2,240.0e+3 ,dx)
r = np.zeros([len(y),len(x)])


for i in range(0,len(x)):
    for j in range(0,len(y)):
        r[j,i] = topography(x[i],y[j])
        
plt.figure()
plt.contour(x,y,r,60)
```

```python
from amrfile import io as amrio
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as col


def addgl(figure,filename,lev,scale,color):
    rhoi=900.0
    rhoo=1000.0
    amr = amrio.load(filename)
    lo,hi = amrio.queryDomainCorners(amr,lev)
    x,y,h = amrio.readBox2D(amr,lev,lo,hi,'thickness',1)
    x,y,r = amrio.readBox2D(amr,lev,lo,hi,'Z_base',1)
    hab = h + rhoo / rhoi * r
    
    time = amrio.queryTime(amr)
    amrio.free(amr)
    plt.contour(x/scale,y/scale,hab, [0], fig=figure ,colors=color)
    print 'time = ',time,' lev = ',lev
    
    
    

def addfl(figure,filename,lev,scale,color):
    rhoi=900.0
    rhoo=1000.0
    amr = amrio.load(filename)
    lo,hi = amrio.queryDomainCorners(amr,lev)
    lo[1] = (hi[1]+1) / 2
    hi[1] = lo[1]+1
    
    x,y,s = amrio.readBox2D(amr,lev,lo,hi,'Z_surface',0)
    x,y,b = amrio.readBox2D(amr,lev,lo,hi,'Z_bottom',0)
    x,y,r = amrio.readBox2D(amr,lev,lo,hi,'Z_base',0)
    
    
    time = amrio.queryTime(amr)
    amrio.free(amr)
    plt.plot(x/scale, s[0,:],color=color)
    plt.plot(x/scale, b[0,:],color=color)
    plt.plot(x/scale, r[0,:],color=color)
    print 'time = ',time,' lev = ',lev
    

#amrio.freeAll()    
    
f = plt.figure()
scale = 1.0e+3
lev=5

addfl(f,'data/MISMIPNG/lazyspinup/plot.ghgChannel.l1l2.5lev.Chombo.090000.2d.hdf5',lev,scale,'red')

addfl(f,'data/MISMIPNG/lazyspinup/plot.ghgChannel.l1l2.5lev.Chombo.110000.2d.hdf5',lev,scale,'pink')

addfl(f,'data/MISMIPNG/lazyspinup/plot.ghgChannel.l1l2.5lev.Chombo.130000.2d.hdf5',lev,scale,'purple')

addfl(f,'data/MISMIPNG/lazyspinup/plot.ghgChannel.l1l2.5lev.Chombo.150000.2d.hdf5',lev,scale,'orange')


plt.axis([1150,1180,-1200,500])
```
