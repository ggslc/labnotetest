```python
from amrfile import io as amrio
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as col

amrio.freeAll()

rhoo=1028.0
rhoi=918.0

#color plot limits
divuhlim = col.Normalize(-150,150)
umislim = col.Normalize(-500,500)
dhlim = col.Normalize(-300,300)
topglim = col.Normalize(-1500,0)


#subregion we care about
lev = 2
lo = np.array([208,624])/4 * 2**lev 
hi = lo + np.array([64,64])/4 * 2**lev 

#fast flow mask
amr = amrio.load('data/plot.ase-testB.4lev.000000.2d.hdf5')
x,y,u = amrio.readBox2D(amr,lev,lo,hi,'xVel',1)
x,y,v = amrio.readBox2D(amr,lev,lo,hi,'yVel',1)
x,y,topg = amrio.readBox2D(amr,lev,lo,hi,'Z_base',1)
x,y,lsrf = amrio.readBox2D(amr,lev,lo,hi,'Z_bottom',1)
x,y,usrf = amrio.readBox2D(amr,lev,lo,hi,'Z_surface',1)
amrio.free(amr)
uu = (u**2 + v**2)**0.5

mask = np.where(uu > 0.0,1.0,0.0)

topga = sum( (topg*mask).transpose())/sum(mask.transpose())
usrfa = sum( (usrf*mask).transpose())/sum(mask.transpose())
lsrfa = sum( (lsrf*mask).transpose())/sum(mask.transpose())

def groundedarea(filename,sp):
    amr = amrio.load(filename)


    #thickness/topography data
    x,y,h = amrio.readBox2D(amr,lev,lo,hi,'thickness',1)
    x,y,r = amrio.readBox2D(amr,lev,lo,hi,'Z_base',1)
    amrio.free(amr)

    scale = 1000.0
    dx = (x[1] - x[0]) / scale
    hab = h + rhoo / rhoi * r

    gm = np.where(hab > 1, 1, 0) * mask
    
    if (sp > 0):
        plt.subplot(sp)
        plt.pcolormesh(x/scale,y/scale,gm)
        plt.colorbar()
        plt.contour(x/1.0e3,y/1.0e3,hab, [1] )
        plt.title(r'r')
        plt.xlabel("x (km)")
        plt.xlabel("y (km)")
        
    ga = sum(gm.transpose())
    return sum(np.where(ga > 32,0,1))
```

```python
g = np.zeros([5])

g[0] = groundedarea("data/plot.ase-testB.4lev.000000.2d.hdf5",221)
g[1] = groundedarea("data/plot.ase-testB.4lev.000061.2d.hdf5",0)
g[2] = groundedarea("data/plot.ase-testB.4lev.000122.2d.hdf5",222)
g[3] = groundedarea("data/plot.ase-testB.4lev.000186.2d.hdf5",223)
g[4] = groundedarea("data/plot.ase-testB.4lev.000258.2d.hdf5",224)

```

```python

p = plt.subplot(211)

t = np.arange(0,5)
plt.plot(g,t)
plt.xlim(0,64)

p = plt.subplot(212)

plt.plot(topga)
plt.plot(usrfa)
plt.plot(lsrfa)
```
