```python
from amrfile import io as amrio
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as col
from matplotlib.backends.backend_pdf import PdfPages
amrio.freeAll()

rhoo=1028.0
rhoi=918.0
scale = 1000.0
#color plot limits
divuhlim = col.Normalize(-150,150)
umislim = col.Normalize(-500,500)
dhlim = col.Normalize(-300,300)
topglim = col.Normalize(-100,100)

def plog(x):
    return np.log10(np.where(x > 1.0e-10,x,1.0e-10))
```

```python
#uniform 16 km mesh
fnam = 'data/BISICLES-Antarctica2013/freshwater/hydrology99.2d.hdf5'
amr = amrio.load(fnam)
lev = 0
lo,hi = amrio.queryDomainCorners(amr,lev)
x,y,Neff = amrio.readBox2D(amr,lev,lo,hi,'Neff',0)
x,y,phi = amrio.readBox2D(amr,lev,lo,hi,'phi',0)
x,y,psi = amrio.readBox2D(amr,lev,lo,hi,'psi',0)
x,y,depth = amrio.readBox2D(amr,lev,lo,hi,'depth',0)
x,y,melt = amrio.readBox2D(amr,lev,lo,hi,'melt',0)
x,y,dux = amrio.readBox2D(amr,lev,lo,hi,'xFlux',0)
x,y,duy = amrio.readBox2D(amr,lev,lo,hi,'yFlux',0)

amrio.free(amr)

```

```python
X,Y = np.meshgrid( np.linspace(min(x),max(x),len(x)),np.linspace(min(x),max(x),len(x)))
```

```python
plt.figure(figsize=(6,5),dpi=300)
Nnorm = col.Normalize(3,7)
plt.pcolormesh(x/scale,y/scale,plog(Neff),norm=Nnorm)
plt.xlim([0,6100])
plt.ylim([0,6100])
plt.xlabel("x (km)")
plt.ylabel("y (km)")
cb = plt.colorbar()
cb.set_label(r'$\log(N_{\rm eff})$ (Pa)')

plt.savefig("neff.png",dpi=300)
```
