```python
from amrfile import io as amrio
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as col

amrio.freeAll()

rhoo = 1028.0
rhoi = 918.0
kice = 2.1 * (3600.0 * 24.0 * 365.25) # ice conductivity, J / (m a K)
cT = 9.76e-8
g=9.81

f='data/BISICLES-Antarctica2013/control/ase/mass-balance/test12/ase-test12-ctrl-2lev-outer.000128.2d.hdf5'
ctrl = amrio.load(f)
level = 0 # 4km resolution
n = 4
lo,hi = amrio.queryDomainCorners(ctrl,level)

lo[0] = 128/n
lo[1] = 600/n
hi[0] = lo[0] + 256/n
hi[1] = lo[1] + 256/n

x,y,h = amrio.readBox2D(ctrl,level,lo,hi,'thck',1)
x,y,s = amrio.readBox2D(ctrl,level,lo,hi,'usrf',1)
x,y,C = amrio.readBox2D(ctrl,level,lo,hi,'Cwshelf',1)
x,y,ub = amrio.readBox2D(ctrl,level,lo,hi,'xVelb',1)
x,y,vb = amrio.readBox2D(ctrl,level,lo,hi,'yVelb',1)
amrio.free(ctrl)


f='data/BISICLES-Antarctica2013/preprocess/ase-temperature-4km.2d.hdf5'
temp=amrio.load(f)
xp,yp,Ts = amrio.readBox2D(ctrl,level,lo,hi,'temp000000',1)
amrio.free(temp)



modu = np.sqrt(ub*ub+vb*vb)
uTb = (modu) * (modu) * C
dT = h * uTb / kice 
T = Ts + dT

km = 1000.0
x = (x -min(x) -1831500.0)/km
y = (y -min(y) -903500.0)/km

#dTlim = col.Normalize(0,20)
#plt.pcolormesh(x,y,dT,norm=dTlim)
Tlim = col.Normalize(253.16,273.16)
Tlim = col.Normalize(-20,0)
mp = 273.16 - h*cT*rhoi*g
plt.pcolormesh(x,y,T-mp,norm=Tlim)
plt.colorbar()

co = plt.contour(x,y,modu/km,np.array([10.0,50.0,100.0,500.0,1000.0])/km,colors='black')
plt.clabel(co,inline=1)
    
plt.savefig("pig_bed_temp_estimate.pdf")

```
